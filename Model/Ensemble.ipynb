```python
# Step 1: Loading and Preparing the Core PCOS Dataset for SVM Modeling

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Step 1: Load and Prepare Core PCOS Dataset for Ensemble Learning (Random Forest)
# Load Core PCOS Dataset
core_pcos_df = pd.read_csv("Core_PCOS_Dataset_Cleaned.csv")

# Clean column names
core_pcos_df.columns = core_pcos_df.columns.str.strip()

# Select numeric features and target
selected_features = [
    "Age (yrs)", "BMI", "hair growth(Y/N)", "Hair loss(Y/N)", "Pimples(Y/N)"
]
target_column = "PCOS (Y/N)"

X_rf = core_pcos_df[selected_features].copy()
y_rf = core_pcos_df[target_column].astype(int)

# Drop any rows with missing values
X_rf = X_rf.dropna()
y_rf = y_rf.loc[X_rf.index]

# Train-test split
X_train_rf, X_test_rf, y_train_rf, y_test_rf = train_test_split(
    X_rf, y_rf, test_size=0.2, random_state=42
)

# Output shapes for verification'
X_train_rf.shape, X_test_rf.shape, y_train_rf.shape, y_test_rf.shape
```

```python
# Import necessary libraries
from sklearn.ensemble import RandomForestClassifier

# Initialize and Train Random Forest Model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train_rf, y_train_rf)
```

```python
# Import evaluation metrics
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns

# Make Predictions
y_pred_rf = rf_model.predict(X_test_rf)

# Calculate Accuracy
acc_rf = accuracy_score(y_test_rf, y_pred_rf)

# Generate Confusion Matrix
cm_rf = confusion_matrix(y_test_rf, y_pred_rf)

# Generate Classification Report
cr_rf = classification_report(y_test_rf, y_pred_rf)

# Print Results
print(" Random Forest Classifier Results")
print(f"Accuracy: {acc_rf:.2f}")
print("Confusion Matrix:")
print(cm_rf)
print("\nClassification Report:")
print(cr_rf)

# Feature Importance Visualization
importances = rf_model.feature_importances_
feature_names = X_rf.columns

# Plotting Feature Importances
plt.figure(figsize=(8, 5))
sns.barplot(x=importances, y=feature_names)
plt.title("Feature Importances - Random Forest Classifier")
plt.xlabel("Importance Score")
plt.ylabel("Features")
plt.show()
```

```python
# Step 3: Visualize the Confusion Matrix Separately (Heatmap for Random Forest)

# Set up the confusion matrix heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(cm_rf, annot=True, fmt="d", cmap="Blues",
            xticklabels=["No PCOS", "PCOS"], yticklabels=["No PCOS", "PCOS"])
plt.title("Confusion Matrix - Random Forest Classifier")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```
